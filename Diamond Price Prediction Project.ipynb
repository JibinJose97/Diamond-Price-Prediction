# Machine Learning

```python
data
```

```python
# Importing LabelEncoder to convert string values into integer values
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data["color"]=encoder.fit_transform(data["color"])
data
```

```python
# Droping strings columns cut and clarity
dataset=data.drop(["cut","clarity"],axis=1)
dataset
```

```python
# As we can see the target column is not at the end of the dataset though droping the target column
dataset1=dataset.drop(["price"],axis=1)
dataset1
```

```python
# Dataset converting into DataFrame using (.values)
x=dataset1.iloc[:,:].values
x
```

```python
# Converting target data into DataFrame
y=dataset.iloc[:,5].values
y
```

```python
# Splitting the data for training and testing
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
# Fitting the LinearRegression model into training set
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(xtrain,ytrain)
```

```python
# Predicting LinearRegression for training set
xpred=regression.predict(xtrain)
xpred
```

```python
# Predicting LinearRegression for testing set
ypred=regression.predict(xtest)
ypred
```

```python
# Importing metrics from sklearn for error calculation
from sklearn import metrics
```

```python
# Printing Train Score, Test Score, Mean Absolute Error, Mean Squared Error, Root Mean Squared Error of LinearRegression
print("Train Score",regression.score(xtrain,ytrain))
print("Test Score",regression.score(xtest,ytest))
print("Mean Abosulte Error",metrics.mean_absolute_error(ytest,ypred))
print("Mean Squared Error",metrics.mean_squared_error(ytest,ypred))
print("Root Mean Squared Error",np.sqrt(metrics.mean_squared_error(ytest,ypred)))
```

```python
# Fitting DecisionTreeRegressor in training set
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(xtrain,ytrain)
```

```python
# Predicting DecisionTreeRegressor for training set
xpred1=dtr.predict(xtrain)
xpred1
```

```python
# Predicting DecisionTreeRegressor for test set
ypred1=dtr.predict(xtest)
ypred1
```

```python
# Printing Train Score, Test Score, Mean Absolute Error, Mean Squared Error, Root Mean Squared Error of DecisionTreeRegression 
print("Train Score",dtr.score(xtrain,ytrain))
print("Test Score",dtr.score(xtest,ytest))
print("Mean Absolute Error",metrics.mean_absolute_error(ytest,ypred1))
print("Mean Squared Error",metrics.mean_squared_error(ytest,ypred1))
print("Root Mean Squared Error",np.sqrt(metrics.mean_squared_error(ytest,ypred1)))
```

```python
# Fitting RandomForestRegressor in training set
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(xtrain,ytrain)
```

```python
# Predicting RandomForestRegressor for training set
xpred2=rfr.predict(xtrain)
xpred2
```

```python
# Predicting RandomForestRegressor for test set
ypred2=rfr.predict(xtest)
ypred2
```
